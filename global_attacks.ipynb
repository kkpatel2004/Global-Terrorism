```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\krupa\OneDrive\Desktop\practice\dataset\globalterroris.csv",encoding='ISO-8859-1')
df
```

```python
df.head()
```

```python
df['approxdate'] = df['approxdate'].fillna(0)
df.isnull().sum()
```

```python
df['related'] = df['related'].fillna(0)
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'country_name'})
```

Attack Counts 

```python
attack_counts = df['attacktype1_txt'].value_counts().head(10)
attack_counts
```

```python
# Plotting the attack types using barplot 
plt.figure(figsize=(10, 6))
sns.barplot(x=attack_counts.values, y=attack_counts.index, palette="viridis")
plt.title("Attack Types", fontsize=16)
plt.xlabel("Number of Attacks", fontsize=12)
plt.ylabel("Attack Type", fontsize=12)
plt.show()
```

```python
weapon_counts = df['weaptype1_txt'].value_counts().head(5)  
weapon_counts
```

```python
weapon_counts.plot(kind='pie', labels=weapon_counts.index, color=['skyblue','pink','red','blue','green','yellow'], figsize=(8, 6), autopct='%1.1f%%')
plt.title("Weapon Count", fontsize=16)
plt.show()
```

```python
kill_counts = df['nkill'].value_counts().sum()
kill_counts
```

```python
kills = df['nkill'].value_counts().head(10).index
filtered_df = df[df['nkill'].isin(kills)]
sns.barplot(x="nkill", y="iday", data=filtered_df, hue="nkill", order=kills, palette="autumn")
plt.title("daily kills")
plt.show() 
```

```python
kills = df['nkill'].value_counts().head(10).index
filtered_df = df[df['nkill'].isin(kills)]
sns.barplot(x="nkill", y="imonth", data=filtered_df, hue="nkill", order=kills, palette="viridis")
plt.title("monthly kills")
plt.show() 
```

```python
kills = df['nkill'].value_counts().head(10).index
filtered_df = df[df['nkill'].isin(kills)]
sns.barplot(x="nkill", y="iyear", data=filtered_df, hue="nkill", order=kills, palette="viridis")
plt.title("yearly kills")
plt.show() 
```

```python
df['nkill'].isnull().sum()
```

```python
df['nkill'] = df['nkill'].fillna(0)
df.isnull().sum()
```

```python
df['nwound'] = df['nwound'].fillna(0)
```

```python
df['nwound'].isnull().sum()
```

```python
df['casulity'] = df['nkill'] + df['nwound']
df['casulity']
```

```python
casulity = df['casulity'].value_counts().head(10)
casulity
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="iyear", y="casulity", palette="viridis", data=df)
plt.xlabel('Casuality', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Casuality Count', fontsize=16)
plt.xticks(rotation=45)  
plt.show()
```
